Funções que são utilizados para o tratamento de dados ausentes.

* **isna()**: Retorna valores booleanos indicando os valores que estão/são ausentes
* **notna()**: Negação de isna, retorna True para valores não NA e False para valores NA.
* **dropna()**: filtra os rótulos dos eixos com base em se os valores de cada rótulo tem dados ausentes e adota limites variáveis para a quantidade de dados a ser tolerada.
* **fillna()**: Preenche os dados ausentes com algum valor ou usando um método de interpolação como 'ffill' ou 'bfill'.
* duplicates(): removendo os dados duplicados

# Gráfico informativo da distribuição da quantidade de carros por marca.

```python
agrupado.plot(kind='pie', autopct='%1.1f%%', figsize=(4, 4))
plt.title('Distribuição de Carros por Marca')
plt.ylabel('')  # Remove o rótulo do eixo y
plt.tight_layout()
plt.show()
```

# Gráfico da evolução da média de preço dos carros ao longo dos meses de 2022 (variável de tempo no eixo X).

```python
carros_2022 = carros[carros["year_of_reference"] == 2022]
price_2022 = carros_2022.groupby("month_of_reference")["avg_price_brl"].mean()
month_order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
price_2022 = price_2022.reindex(month_order)
plt.figure(figsize=(10, 6))
price_2022.plot(kind="line", color='black')
plt.xlabel("Mês de Referência")
plt.ylabel("Preço Médio (BRL)")
plt.title("Preço Médio por mês em 2022")
plt.xticks(range(len(price_2022.index)), price_2022.index, rotation=45, ha='right')

for i, price in enumerate(price_2022.values):
    plt.text(i, price, f'{price:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

# Gráfico da distribuição da média de preço dos carros por marca e tipo de engrenagem.

```python
average_price_by_brand_gear = carros.groupby(['brand', 'gear'])['avg_price_brl'].mean().unstack()


plt.figure(figsize=(12, 8))
average_price_by_brand_gear.plot(kind='bar', stacked=False, ax=plt.gca())

plt.xlabel('Marca')
plt.ylabel('Preço Médio (BRL)')
plt.title('Distribuição da Média de Preço dos Carros por Marca e Tipo de Engrenagem')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tipo de Engrenagem', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Percebemos que a média de preços dos carros automaticos são maiores em todas as marcas menos na Renault.

Fiat e VW possuem os carros automaticos com maior média de preço e também os manuais com menor médioa de preço.

A renault é a marca com a menor variação na média de preço entre carros automáticos e manuais.

# Gráfico da distribuição da média de preço dos carros por marca e tipo de combustível

```python
average_price_by_brand_fuel = carros.groupby(['brand', 'fuel'])['avg_price_brl'].mean().unstack()
colors = ["teal","powderblue","darkturquoise"]


plt.figure(figsize=(12, 8))
average_price_by_brand_fuel.plot(kind='bar', stacked=False, ax=plt.gca(), width=0.9,color=colors)

plt.xlabel('Marca')
plt.ylabel('Preço Médio (BRL)')
plt.title('Distribuição da Média de Preço dos Carros por Marca e Tipo de Combustível')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tipo de Combustível', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Os carros a alcool são os mais baratos entre os 3 tipos de combustíveis. Nem a Nissan nem a Renault apresentam carros a alcool nesse database.
Os carros a Disel são os mais caros e a VW tem a maior média nesse modelo.
Dentre os carros a gasolina a Nissan e a GM apresentam a maior média de preço.
